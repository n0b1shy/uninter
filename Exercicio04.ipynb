```
# Mensagem de boas-vindas // Exi 1/8
print("Bem-vindo ao Sistema de Gerenciamento de Livros de Josué Cordeiro")

# Inicialização da lista vazia para armazenar os livros e variável id_global
lista_livros = []  # Exi 2/8 Lista vazia para armazenar dicionários de livros
id_global = 0  # Contador inicial para IDs dos livros

# Função para cadastrar um novo livro
def cadastrar_livro(id):
    # Solicita os dados do livro ao usuário
    nome = input("Digite o nome do livro: ")
    autor = input("Digite o autor do livro: ")
    editora = input("Digite a editora do livro: ")

    # Cria um dicionário com os dados do livro Exi 3/8
    livro = {
        "id": id,
        "nome": nome,
        "autor": autor,
        "editora": editora
    }

    # Adiciona o dicionário à lista de livros
    lista_livros.append(livro)
    print(f"Livro '{nome}' cadastrado com sucesso!")

# Função para consultar livros Exi 4/8
def consultar_livro():
    while True:
        # Menu de opções de consulta
        print("\nMenu Consultar Livro")
        print("1. Consultar Todos")
        print("2. Consultar por Id")
        print("3. Consultar por Autor")
        print("4. Retornar ao menu")

        opcao = input("Escolha uma opção: ")
        # Consulta todos os livros
        if opcao == "1":
            if lista_livros:
                print("\nLista de todos os livros:")
                for livro in lista_livros:
                    print(f"ID: {livro['id']}, Nome: {livro['nome']}, Autor: {livro['autor']}, Editora: {livro['editora']}")
            else:
                print("Nenhum livro cadastrado.")

        # Consulta livro por ID
        elif opcao == "2":
            id_consulta = input("Digite o ID do livro: ")
            encontrado = False
            for livro in lista_livros:
                if str(livro['id']) == id_consulta:
                    print(f"ID: {livro['id']}, Nome: {livro['nome']}, Autor: {livro['autor']}, Editora: {livro['editora']}")
                    encontrado = True
                    break
            if not encontrado:
                print("ID inválido.")

        # Consulta livros por autor
        elif opcao == "3":
            autor_consulta = input("Digite o nome do autor: ")
            encontrados = False
            for livro in lista_livros:
                if livro['autor'].lower() == autor_consulta.lower():
                    print(f"ID: {livro['id']}, Nome: {livro['nome']}, Autor: {livro['autor']}, Editora: {livro['editora']}")
                    encontrados = True
            if not encontrados:
                print("Nenhum livro encontrado para este autor.")

        # Retorna ao menu principal
        elif opcao == "4":
            break

        # Opção inexistente
        else:
            print("Opção inexistente.")

# Função para remover um livro Exi 5/8
def remover_livro():
    while True:
        id_remover = input("Digite o ID do livro a ser removido: ")
        for livro in lista_livros:
            if str(livro['id']) == id_remover:
                lista_livros.remove(livro)
                print(f"Livro com ID {id_remover} foi removido com êxito"!")
                return
        print("ID inexistente.")

# Menu principal
while True:
    # Exibe o menu principal Exi 6/8
    print("\nMenu Principal")
    print("1. Cadastrar Livro")
    print("2. Consultar Livro(s)")
    print("3. Remover Livro")
    print("4. Encerrar programa")

    opcao = input("Escolha uma opção: ")

    # Cadastra um novo livro
    if opcao == "1":
        id_global += 1
        cadastrar_livro(id_global)

    # Consulta livros
    elif opcao == "2":
        consultar_livro()

    # Remove um livro
    elif opcao == "3":
        if lista_livros:
            remover_livro()
        else:
            print("Nenhum livro cadastrado para remover.")

    # Encerra o programa
    elif opcao == "4":
        print("Programa encerrado.")
        break

    # Opção inválida Exi 8/8
    else:
        print("Opção inválida.")
```
